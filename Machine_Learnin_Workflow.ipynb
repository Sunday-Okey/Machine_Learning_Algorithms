Data Collection and Exploration

We'll build a model that predicts whether a patient has breast cancer.

image.png

The scikit-learn library stores several datasets, including
[this breast cancer dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset)

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
#This will return the data attribute as a pandas' DataFrame and the target attribute as a Series.
cancer_data = load_breast_cancer(as_frame=True)
cancer_df = cancer_data.data
cancer_df['target'] = cancer_data.target
```

```python
cancer_df.head()
```

```python
cancer_df['target'].value_counts()
```

```python
cancer_df.shape
```

```python
#get the number of na values in each column
cancer_df.isnull().sum(axis =0)
```

```python
#Get the number of na values in each row
cancer_df.isnull().sum(axis = 1)
```

```python
cancer_df.describe()
```

```python
from sklearn.model_selection import train_test_split
y = cancer_df.drop('target', axis=1)
X = cancer_df.iloc[:, :-1]
```

```python
X
```

```python
y = cancer_df.iloc[:, -1]
y
```

```python
#Since the data is split randomly, setting a random_state to a specific number allows us to
#reproduce the same results every time we run the code.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=417)
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC(penalty="l2",loss="squared_hinge",C=10, random_state=417)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python

test_accuracy = clf.score(X_test, y_test)
test_accuracy 
```

```python
model = LinearSVC(penalty="l2",
loss="squared_hinge",
C=20,
max_iter=3500,
random_state=417)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
